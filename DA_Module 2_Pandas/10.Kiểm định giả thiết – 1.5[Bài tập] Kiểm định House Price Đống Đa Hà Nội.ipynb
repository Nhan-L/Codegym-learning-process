```python
# Sử dụng kiểm định Independent T test với hàm ttest_ind với kiểm định trái
```

```python
df1 = pd.read_csv("Database_practice\house_price_Dống-Da_Hà-Nội_subdata.csv")
df1.head()
```

```python
# Tạo cột triệu đ/m2
df1['triệu đ/m2'] = round(df1['price'] * 1000 / df1['area'],0) 
df1.head()
```

```python
df1 = df1.replace([np.inf, -np.inf], np.nan)
df1.head()
```

```python
df1.fillna({'triệu đ/m2':0}, inplace = True)
df1.head()
```

```python
df1['land_certificate'].unique()
```

```python
df1_sodo = df1.loc[(df1['land_certificate'] == 'So do')]
df1_sodo.head()
```

```python
df1_kosodo = df1[df1['land_certificate'].isna()]
df1_kosodo.head()
```

```python
print(stats.ttest_ind(df1_sodo['triệu đ/m2'], df1_kosodo['triệu đ/m2'], equal_var=False))
```

Đọc kết quả:

- Chúng ta có thể thấy Pvalue = 0.5064 >> 5%, chấp nhận giả thuyết không, bác bỏ giả thuyết đối
- Kết luận: Giá của những căn nhà không có thông tin về giấy tờ pháp lý thấp hơn giá nhà những căn có thông tin về giấy tờ pháp lý với mức ý nghĩa 5%
