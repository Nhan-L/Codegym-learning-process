# 3.Giữa thuộc tính land_certificate và property_type có tương quan với nhau


```python
# Kiểm định Chi bình phương để kiểm định tương quan giữa hai thuộc tính định tính
```

```python
data = df1.groupby('property_type').count()
data
```

```python
chi, p, dof, expected = chi2_contingency (data)
print ('Chi-square value = ', chi)
print ('p value = ', p)
print ('Degree of freedom = ', dof)
print ('expected table = ', expected)
```

* KẾT QUẢ
- H0 không có mối tương quan giữa property_type với các thuộc tính trong đó có land_certificate
- H1 có mối tương quan giữa property_type với các thuộc tính trong đó có land_certificate
- Với pvalue = 1.0 > alpha = 0.05, chấp nhận H0, bác bỏ H1, nên không có mối tương quan giữa property_type với các thuộc tính trong đó có land_certificate

# 4.Hãy  mã hóa lại thuộc tính giá nhà theo đơn vị (nghìn đồng/m2) và sắp xếp giá nhà thành 4 mức tương ứng với các khoảng tứ phân vị. Tiến hành kiểm định tương quan giữa biến giá nhà này với các biến tọa độ vị trí (lat, long)   

```python
# Kiểm định spearman để kiểm định tương quan giữa một thuộc tính thứ bậc và một thuộc tính liên tục
```

```python
# Tạo cột giá nhà nghìn đồng/m2
df1['nghinm2'] = round(df1['price'] * 1000000 / df1['area'],0)
df1.head()
```

```python
df2 = df1[['nghinm2', 'lat', 'long' ]]
df2.head()
```

```python
df3 = df2.replace([np.inf, -np.inf], 0)
df3.head()
```

```python
df3.shape
```

```python
# Lấy gía trị nhỏ nhất của nghinm2
nghinm2s = sorted(df3['nghinm2'].unique(), reverse = True)
rank_nghinm2 = {nghinm2s[i]:i+1 for i in range (len(nghinm2s))}
rank_nghinm2
```

```python
rank = pd.Series(rank_nghinm2)
rank
```

```python
h = df3['nghinm2']
df3['nghinm2_Rank'] = h.map(rank)
df3
```

```python
# Biến nghinm2_Rank với lat
r, pvalue = stats.spearmanr(df3['nghinm2_Rank'], df3['lat'])
print('r = ', r)
print('pvalue = ', pvalue)
```

Kết luận:
- Do p_value (0.20) > alpha = 0.05 nên H0 được chấp nhận, HA bị loại bỏ, giữa các thuộc tính không có tương quan, r gần bằng 0, mối quan hệ giữa 2 thuộc tính không chặt chẽ

```python
# Biến nghinm2_Rank với long
r, pvalue = stats.spearmanr(df3['nghinm2_Rank'], df3['long'])
print('r = ', r)
print('pvalue = ', pvalue)
```

Kết luận:

- Do p_value (7.821453782581783e-09) < alpha = 0.05 nên H0 bị loại bỏ, HA được chấp nhận, giữa các thuộc tính có tương quan, r > 0, mối quan hệ giữa 2 thuộc tính không chặt chẽ lắm
