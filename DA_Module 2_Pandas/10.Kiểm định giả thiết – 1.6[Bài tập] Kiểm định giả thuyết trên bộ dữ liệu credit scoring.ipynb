```python
import pandas as pd
from scipy import  stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Database_practice\Credit_Scoring.csv')
df.head()
```

# Bạn hãy thực hiện những kiểm định sau để phân tích đặc điểm của khách hàng tiềm năng

# 1.Có phải những khách hàng không có người phụ thuộc sẽ có thu nhập trung bình theo tháng (MonthlyIncome) nhỏ hơn những khách hàng có người phụ thuộc không (với mức ý nghĩa 10%)


```python
# Sử dụng kiểm định Independent T test với hàm ttest_ind với kiểm định cho kiểm định 1 phía (trái hay phải tùy theo cách bạn viết giả thuyết)
```

```python
df['NumberOfDependents'].unique()
```

```python
df.fillna({'MonthlyIncome':0}, inplace = True)
df.head()
```

```python
df_coNPT = df[df['NumberOfDependents'] > 0]
df_coNPT.head()
```

```python
df_koNPT = df[df['NumberOfDependents'] == 0]
df_koNPT.head()
```

```python
print(stats.ttest_ind(df_coNPT['MonthlyIncome'], df_koNPT['MonthlyIncome'], equal_var=False))
```

Đọc kết quả:

- Chúng ta có thể thấy Pvalue = 6.8070257779524515e-248 << 10%, bác bỏ giả thuyết không, chấp nhận giả thuyết đối
- Kết luận: những khách hàng không có người phụ thuộc sẽ có thu nhập trung bình theo tháng (MonthlyIncome) lớn hơn những khách hàng có người phụ thuộc không (với mức ý nghĩa 10%)

# 2.Có phải trung bình số lượng khoản vay (NumberOfOpenCreditLinesAndLoans) những khách hàng gặp khó khăn trong vòng 2 năm trở lại đây (SeriousDlqin2yrs =1) thì sẽ cao hơn những khách hàng không gặp khó khăn không với mức ý nghĩa 10%  

```python
# Sử dụng kiểm định Independent T test với hàm ttest_ind với kiểm định cho kiểm định 1 phía (trái hay phải tùy theo cách bạn viết giả thuyết)
```

```python
df['NumberOfOpenCreditLinesAndLoans'].unique()
```

```python
df['SeriousDlqin2yrs'].unique()
```

```python
df_gapkhokhan = df[df['SeriousDlqin2yrs'] == 1]
df_gapkhokhan.head()
```

```python
df_kogapkhokhan = df[df['SeriousDlqin2yrs'] == 0]
df_kogapkhokhan.head()
```

```python
print(stats.ttest_ind(df_gapkhokhan['NumberOfOpenCreditLinesAndLoans'], df_kogapkhokhan['NumberOfOpenCreditLinesAndLoans'], equal_var=False))
```

Đọc kết quả:

- Chúng ta có thể thấy Pvalue = 7.875696017574721e-26 << 10%, bác bỏ giả thuyết không, chấp nhận giả thuyết đối
- Kết luận:  trung bình số lượng khoản vay (NumberOfOpenCreditLinesAndLoans) những khách hàng gặp khó khăn trong vòng 2 năm trở lại đây (SeriousDlqin2yrs =1) thì sẽ thấp hơn những khách hàng không gặp khó khăn không với mức ý nghĩa 10%
