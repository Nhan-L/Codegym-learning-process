# 4.Mã hóa lại thuộc tính MonthlyIncome thành thuộc tính MonthlyIncome_order theo các khoảng tứ phân vị, giữa thuộc tính mới này và tình trạng nợ xấu trong 2 năm trở lại đây (SeriousDlqin2yrs) có liên quan tới nhau không 

```python
# Kiểm định Chi bình phương để kiểm định tương quan giữa hai thuộc tính định tính
```

```python
# Lấy gía trị nhỏ nhất của MonthlyIncome
MonthlyIncomes = sorted(df['MonthlyIncome'].unique(), reverse = True)
rank_MonthlyIncome = {MonthlyIncomes[i]:i+1 for i in range (len(MonthlyIncomes))}
rank_MonthlyIncome
```

```python
rank = pd.Series(rank_MonthlyIncome)
rank
```

```python
h = df['MonthlyIncome']
df['MonthlyIncome_order'] = h.map(rank)
df
```

```python
# Biến MonthlyIncome_order với SeriousDlqin2yrs
r, pvalue = stats.spearmanr(df['MonthlyIncome_order'], df['SeriousDlqin2yrs'])
print('r = ', r)
print('pvalue = ', pvalue)
```

Kết luận:

- Do p_value (2.1433575685353356e-29) < alpha = 0.05 nên H0 bị loại bỏ, HA được chấp nhận, giữa thuộc tính MonthlyIncome_order và SeriousDlqin2yrs có tương quan, r gần bằng 0, mối quan hệ giữa 2 thuộc tính không chặt chẽ

# 5.Giữa thuộc tính MonthlyIncome_order với thuộc tính tỉ lệ số dư tài khoản (RevolvingUtilizationOfUnsecuredLines) có mối liên hệ với nhau không    

```python
# Kiểm định spearman để kiểm định tương quan giữa một thuộc tính thứ bậc và một thuộc tính liên tục
```

```python
# Biến MonthlyIncome_order với SeriousDlqin2yrs
r, pvalue = stats.spearmanr(df['MonthlyIncome_order'], df['RevolvingUtilizationOfUnsecuredLines'])
print('r = ', r)
print('pvalue = ', pvalue)
```

Kết luận:

- Do p_value (1.3872297271959383e-10) < alpha = 0.05 nên H0 bị loại bỏ, HA được chấp nhận, giữa thuộc tính MonthlyIncome_order và RevolvingUtilizationOfUnsecuredLines có tương quan, r < 0, mối quan hệ giữa 2 thuộc tính không chặt chẽ và có thẻ coi như không liên quan
